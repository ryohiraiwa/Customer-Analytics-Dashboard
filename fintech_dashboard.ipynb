```python
import pandas as pd
import numpy as np
from faker import Faker
from datetime import datetime

fake = Faker()
np.random.seed(42)

n_customers = 10_000

countries = ["UK", "UAE", "Spain", "Portugal", "Poland", "France", "Germany", "Netherlands"]
plans = ["Standard", "Plus", "Premium", "Metal"]

customers = pd.DataFrame({
    "customer_id": range(1, n_customers + 1),
    "name": [fake.name() for _ in range(n_customers)],
    "country": np.random.choice(countries, n_customers, p=[0.35, 0.12, 0.12, 0.08, 0.10, 0.08, 0.10, 0.05]),
    "age": np.random.randint(18, 65, n_customers),
    "signup_date": pd.to_datetime([
        fake.date_between(start_date="-24M", end_date="today") for _ in range(n_customers)
    ]),
    "plan": np.random.choice(plans, n_customers, p=[0.58, 0.18, 0.16, 0.08]),
})

customers["is_paid_plan"] = customers["plan"].isin(["Plus", "Premium", "Metal"])

customers.head()
```
